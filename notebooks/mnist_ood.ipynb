```python
from models.score_base import VisionScoreModel
from configs import mnist_config as cfg

config = cfg.get_config()
workdir = "workdir/mnist/resnext_v0/"

ckpt_dir = os.path.join(workdir,"checkpoints", "last.ckpt")
# ckpt_dir = os.path.join(workdir,"checkpoints-meta", "last.ckpt")
scorenet = VisionScoreModel.load_from_checkpoint(
    checkpoint_path=ckpt_dir, config=config).cuda()

scorenet.eval()
```

```python
%%time
config.data.dataset = "mnist"
mnist_train, mnist_test = get_dataset(config, mode="test")
config.data.dataset = "fashion"
fashion_train, fashion_test = get_dataset(config, mode="test")
config.data.dataset = "omniglot"
omni_train, omni_test = get_dataset(config, mode="test")
```

```python
%%time

score_norms = []
labels = []

for x_batch, y in mnist_test:
    s = scorenet.scorer(x_batch.cuda(), denoise_step=True).cpu().numpy()
    score_norms.append(s)    
    labels.append(0)
    
len(score_norms)
```

```python
%%time

ood_score_norms = {}
ood_labels = {}

lab=0
for name, ood_data in [("fashion", fashion_test), ("omni", omni_test)]:
    lab += 1
    ood_score_norms[name] = []
    ood_labels[name] = []
    for x_batch, y in ood_data:
        s = scorenet.scorer(x_batch.cuda(), denoise_step=True).cpu().numpy()
        ood_score_norms[name].append(s)
        ood_labels[name].append(lab)        
```

```python
scores = np.vstack(score_norms)
label_arr = np.zeros(scores.shape[0])
scores.shape
```

```python
fashion_ood_scores = np.vstack(ood_score_norms["fashion"])
fashion_ood_labels = np.zeros(fashion_ood_scores.shape[0]) + 1

omni_ood_scores = np.vstack(ood_score_norms["omni"])
omni_ood_labels = np.zeros(omni_ood_scores.shape[0]) + 2
```

```python
np.concatenate((fashion_ood_scores, omni_ood_scores)).shape, omni_ood_scores.shape, fashion_ood_scores.shape
```

```python
df = pd.DataFrame(np.concatenate((scores, fashion_ood_scores, omni_ood_scores)),
                    columns=[f"t{i}" for i in range(config.model.num_scales)])
df["labels"] = np.concatenate((label_arr, fashion_ood_labels, omni_ood_labels))
df
```

```python
df_melt = df.melt(id_vars="labels", var_name="tau")
df_melt.head()
```
